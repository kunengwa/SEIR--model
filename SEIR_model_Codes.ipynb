```python
#importing a few important libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# parameter and parameters values
N = 10000
mu = 1/(64*365)
pi = N*mu
alpha =0.00357
sigma =1/7
beta =0.015
delta =0.0002
rho_s =0.001
rho_b =0.002
nu = 0.005
phi = 0.3

# initial number of infected and recovered individuals
s_initial = 10000.00
sb_initial = 0.00
e_initial = 0.00
i_initial = 100
r_initial = 0.00
C_initial = 100
#defining a function that solves the differential equations 
def deriv(x, t, mu, pi, alpha, sigma, beta, delta, rho_s, rho_b, nu, phi):
#Assigning list of state variables to x
  s, sb, e, i, r, C= x
  #defining the ODEs
  dsdt = pi-mu*s-rho_b*s-(nu*i*s)/(s+sb+e+i+r)-(beta*i*s)/(s+sb+e+i+r)+rho_s*sb
  dsbdt = rho_b*s+(nu*i*s)/(s+sb+e+i+r)-(mu+rho_s)*sb-(phi*beta*i*sb)/(s+sb+e+i+r)
  dedt = (beta*i*s)/(s+sb+e+i+r)+(phi*beta*i*sb)/(s+sb+e+i+r)-(mu+sigma)*e
  didt = sigma*e-(mu+alpha+delta)*i
  drdt = alpha*i-mu*r
  dCdt = sigma*e
  #returning the solution of each differential equations
  return [dsdt, dsbdt, dedt, didt, drdt, dCdt]
#specifying time frame
t = np.linspace(0,1500, 1500)
#Asigning a the initials conditions to the variables x_initials conditions
x_initial = s_initial, sb_initial, e_initial, i_initial, r_initial, C_initial
#Using the command odeint to actually solve the differential equations
soln = odeint(deriv, x_initial, t, args=(mu, pi, alpha, sigma, beta, delta, rho_s, rho_b, nu, phi))
#Assigning the each particularly solution to each state variables
s, sb, e, i, r, C = soln.T 
#Creating a list of labels and picking a differnt each iteration
Labels=[r'$\rho_s=0$',r'$\rho_s=0.0005$',r'$\rho_s=0.001$',r'$\rho_s=0.0015$']
#Picking a different color each iteration
```
